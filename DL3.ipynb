# Implementation of Perceptron Learning

## Introduction

The **Perceptron** is one of the earliest and simplest types of artificial neural networks used in machine learning. It is a **supervised learning algorithm** designed for **binary classification problems**, where the output belongs to one of two classes.

The perceptron learns a **linear decision boundary** that separates the data points of different classes. It works by adjusting its weights based on the error between the predicted output and the actual output during training.

---

## Perceptron Model

A perceptron consists of the following components:

- **Input Features (x₁, x₂, ..., xₙ)** – The input variables of the dataset.
- **Weights (w₁, w₂, ..., wₙ)** – Parameters that determine the importance of each input.
- **Bias (b)** – A constant value added to shift the decision boundary.
- **Activation Function** – Determines the output of the neuron.

The perceptron calculates the weighted sum of the inputs and applies a **sign activation function** to produce the output.

### Mathematical Representation

The output of a perceptron can be represented as:

y = sign(w₁x₁ + w₂x₂ + ... + wₙxₙ + b)

where:
- **xᵢ** represents the input features
- **wᵢ** represents the weights
- **b** represents the bias
- **y** represents the predicted class label

The sign function returns:

- **+1** for the positive class
- **-1** for the negative class

---

## Perceptron Learning Algorithm

The perceptron learns by **iteratively adjusting the weights** whenever the predicted output differs from the actual output.

### Weight Update Rule

The weight update rule is given by:

w(new) = w(old) + η * y * x

where:
- **η (eta)** is the learning rate
- **y** is the true label
- **x** is the input feature vector

Similarly, the bias is updated as:

b(new) = b(old) + η * y

These updates move the decision boundary so that the classifier correctly separates the training data.

---

## Working of the Algorithm

The perceptron learning algorithm follows these steps:

1. Initialize weights and bias with small values (often zeros).
2. For each training example:
   - Compute the weighted sum of inputs.
   - Apply the activation function to get the predicted output.
3. Compare the predicted output with the actual label.
4. If the prediction is incorrect, update the weights and bias using the update rule.
5. Repeat the process for multiple iterations (epochs) until the model converges.

---

## Advantages of Perceptron

- Simple and easy to implement
- Computationally efficient
- Works well for **linearly separable data**
- Forms the basis for modern neural networks

---

## Limitations

- Can only solve **linearly separable problems**
- Cannot solve problems like the **XOR problem**
- Limited to **binary classification**

---

## Conclusion

The perceptron is a foundational algorithm in machine learning and neural networks. Although simple, it introduced the concept of learning weights from data and laid the groundwork for more advanced models such as **multilayer perceptrons and deep neural networks**.

```python
import numpy as np

# Training dataset (AND gate example)
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([-1,-1,-1,1])   # Labels

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10

# Perceptron training
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = 1 if linear_output > 0 else -1

        # Update rule
        if prediction != y[i]:
            weights = weights + learning_rate * y[i] * X[i]
            bias = bias + learning_rate * y[i]

print("Final Weights:", weights)
print("Final Bias:", bias)
```

```python
def predict(x):
    result = np.dot(x, weights) + bias
    return 1 if result > 0 else -1

for sample in X:
    print(sample, "->", predict(sample))
```
